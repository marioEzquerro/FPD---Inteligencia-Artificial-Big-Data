Importo librerías necesarias para crear estructuras que mostrar

```python
# primero importaremos la librería de panda
import pandas as pd

# habrá ejemplos de uso de la librería numPy
import numpy as np
```

### Visualización de la información a través de MatplotLib

```python
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(5))
print(ts)
```

```python
# se puede visualizar una Serie de forma simple

ts.plot()
plt.grid(True) #configurar el plot
plt.show()     #mostrar la configuración
```

#### Diagramas de barras

```python
# también se puede visualizar un Dataframe de una forma muy simple

df1 = pd.DataFrame(np.random.randn(10, 4), columns=['A', 'B', 'C', 'D'])
print(df1)

df1.plot.bar()  #diagrama de barras vertical
plt.grid(True)
plt.show()

print()

df1.plot.barh()  #diagrama de barras horizontal
plt.grid(True)
plt.show()

print()

df2 = pd.DataFrame(np.random.rand(10, 4), columns=['A', 'B','C', 'D'])
print(df2)
df2.plot.bar(stacked=True)  #se puede mostrar en barras apiladas
plt.grid(True)
plt.show()
```

```python
# existen datos que se ven mejor en forma de área
df2.plot.area()
plt.grid(True)
plt.show()

print()

# existen datos que se ven mejor en forma de área (también apilada)
df2.plot.area(stacked=False)
plt.grid(True)
plt.show()
```

#### Mapas de calor

```python
# muchas veces es muy util un heatmap
import seaborn as sns

sns.heatmap(df2, annot=True)
```

#### Diagrama de quesos

```python
# si deseamos realizar un diagrama de quesos

df3 = pd.DataFrame(np.random.rand(4), index=['A', 'B', 'C','D'])
print(df3)

df3.plot.pie(subplots=True)
plt.show()
```
