```python
from transformers import pipeline
from transformers import Conversation
```

Preparamos los pipelines a los modelos escogidos


```python
offensive_filter = pipeline(task="text-classification", model="cardiffnlp/twitter-roberta-base-sentiment")
offensive_filter_2 = pipeline(task="text-classification", model="Seethal/sentiment_analysis_generic_dataset")
chat_bot = pipeline(task="conversational", model="facebook/blenderbot-400M-distill")
```

### Comprobaciones en el chat
Aqui ambos modelos evaluan que tipo de mensaje es (`0 -> Negativo; 1 -> Neutral; 2 -> Positivo`). Comprobamos que ambos dan `LABEL_0` osea, el agente detecta que el mensaje es negativo y luego verificamos que la tasa sea >70%.

Tambien he decidido añadir un pipeline al modelo de conversaciones encontrado en los apuntes de [HuggingFace](https://colab.research.google.com/drive/13R35kF3jdSI24y7NmyiXNfNbnmoe6uzP?authuser=2#scrollTo=UClFi5m4legyZ)




```python
responses = list()

# Ver si la valoracion del modelo es de mesnaje ofensivo
def detect_label(resp):
  return True if resp[0]['label'] == 'LABEL_0' else False

# Obtener %
def get_score(resp):
  return resp[0]['score']

# Validar tipo de mensaje y % de ambos modelos
def is_sentence_ofensive(usr_input):
  result_1 = offensive_filter(usr_input)
  result_2 = offensive_filter_2(usr_input)
  return True if detect_label(result_1) and detect_label(result_2) and get_score(result_1) > .7 and get_score(result_2) > .7 else False
```

### Bucle del chat
Aqui el programa se ejecuta en bucle hasta que el usuario introduzca 'q' por teclado.

```python
print('--- CHATBOT ---')
print('Press "q" to exit\n')

while (True):
  usr_input = input('You: ')

  if usr_input == 'q':
    break

  if is_sentence_ofensive(usr_input):
      print('❗ Abusive language will not be tolerated ❗') 
  else:
    conversation = Conversation(usr_input, generated_responses=responses)
    chat_bot([conversation])
    conversation.mark_processed()
    print(f'🤖: {responses[-1]}')

print('🤖: BYE!')
print('--- END ---')  
```
