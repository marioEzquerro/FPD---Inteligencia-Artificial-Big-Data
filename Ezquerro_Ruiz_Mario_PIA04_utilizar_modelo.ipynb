	* Longitud sépalo (cm.)		
	* Anchura sépalo (cm.)
	* Longitud pétalo (cm.)
	* Anchura pétalo (cm.)
2. El programa devolverá si es iris setosa, iris versicolor, o iris virginica; junto con la probabilidad de que lo sea.

*Al crear otro fichero .ipynb deberás subir el fichero “modelo_ud04.h5” a Google Collab y posteriormente importarlo con el código que te indique ChatGPT.*

```python
import numpy as np

# Pedimos al usuario los valores
sepal_length = float(input("Introduce la longitud del sépalo en cm.: "))
sepal_width = float(input("Introduce la anchura del sépalo en cm.: "))
petal_length = float(input("Introduce la longitud del pétalo en cm.: "))
petal_width = float(input("Introduce la anchura del pétalo en cm.: "))

# Convertimos los valores introducidos por el usuario en un arreglo de NumPy
input_data = np.array([[sepal_length, sepal_width, petal_length, petal_width]])

# Realizamos la predicción sobre los datos introducidos por el usuario
prediction = loaded_model.predict(input_data)

# Mostramos la predicción
if prediction[0,0] > prediction[0,1]:
    if prediction[0,0] > prediction[0,2]:
        print("La flor es de la especie setosa con un", prediction[0,0]*100, "% de confianza")
    else:
        print("La flor es de la especie virginica con un", prediction[0,2]*100, "% de confianza")
else:
    if prediction[0,1] > prediction[0,2]:
        print("La flor es de la especie versicolor con un", prediction[0,1]*100, "% de confianza")
    else:
        print("La flor es de la especie virginica con un", prediction[0,2]*100, "% de confianza")
```
